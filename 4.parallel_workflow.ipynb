```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class BatsmanState(TypedDict):
    runs: int
    balls: int
    fours: int
    sixes: int

    sr: float
    bpb: float
    boundary_percent: float
```

```python
def strike_rate(state: BatsmanState) :
    runs   = state['runs']
    balls  = state['balls']
    fours  = state['fours']
    sixes  = state['sixes']

    sr = (runs / balls) * 100

    return {'sr':sr}
```

```python
def ball_per_boundary(state: BatsmanState) :
    runs   = state['runs']
    balls  = state['balls']
    fours  = state['fours']
    sixes  = state['sixes']

    bpb = (balls / (fours + sixes)) 

    return {'bpb':bpb}
```

```python
def boundary_percentage(state: BatsmanState) :
    runs   = state['runs']
    balls  = state['balls']
    fours  = state['fours']
    sixes  = state['sixes']

    boundary_percent = (((4 * fours) + (6 * sixes))/runs) * 100 

    return {'boundary_percent': boundary_percent}
```

```python
graph = StateGraph(BatsmanState)

# make nodes in graph

graph.add_node('strike_rate', strike_rate)
graph.add_node('ball_per_boundary', ball_per_boundary)
graph.add_node('boundary_percentage', boundary_percentage)


# make edges in graph

graph.add_edge(START, 'strike_rate')
graph.add_edge(START, 'ball_per_boundary')
graph.add_edge(START, 'boundary_percentage')

graph.add_edge('strike_rate', END)
graph.add_edge('ball_per_boundary', END)
graph.add_edge('boundary_percentage', END)

# compile the graph

workflow = graph.compile()

```

```python
# execute the graph

initial_state = {'runs':100, 'balls':100, 'fours':10, 'sixes': 5}

final_state  = workflow.invoke(initial_state)
```

```python
strike_rate  = final_state['sr']
bpb  = final_state['bpb']
boundary_percent  = final_state['boundary_percent']


print(f'Strike rate : {strike_rate} ,\nballs per boundary : {bpb} ,\nboundary_percent: {boundary_percent}')
```
