```python
from langgraph.graph import StateGraph, START, END
from langchain_ollama import ChatOllama

from typing import TypedDict
```

```python
llm = ChatOllama(model='llama3.2:3b')
```

```python
# create state

class BlogState(TypedDict):
    topic: str
    outline: str
    blog: str
```

```python
def generate_outine(state:BlogState) -> BlogState:

    # extract the topic
    topic = state['topic']

    # create a prompt
    prompt = f'Generate a detailed outline on the topic {topic}'

    # call the llm

    outline = llm.invoke(prompt).content

    # update the state
    state['outline']= outline

    return state
```

```python
def generate_blog(state:BlogState) -> BlogState:

    # extract the topic
    topic = state['topic']
    outline = state['outline']

    # create a prompt
    prompt = f'Generate a blog on the topic {topic} based upon the detailed outline given below {outline}'

    # call the llm

    blog = llm.invoke(prompt).content

    # update the state
    state['blog']= blog

    return state
```

```python
graph = StateGraph(BlogState)

# add node
graph.add_node('generate_outine',generate_outine)
graph.add_node('generate_blog',generate_blog)

# add edge
graph.add_edge(START, 'generate_outine')
graph.add_edge('generate_outine', 'generate_blog')
graph.add_edge('generate_blog', END)

# compile the graph

workflow = graph.compile()
```

```python
initial_state = {'topic':'Rise & Fall of IPL'}
final_state = workflow.invoke(initial_state)

outline = final_state['outline']

blog = final_state['blog']

print(outline)
print(blog)
```
